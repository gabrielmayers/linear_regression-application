```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Loading Data

```python
dataset = pd.read_csv('EcommerceCustomers')
```

```python
dataset.head()
```

## Exploring the Data

```python
sns.pairplot(dataset)
plt.savefig('Desktop')
```

```python
sns.lmplot(x='Length of Membership', y='Yearly Amount Spent', data=dataset)
plt.savefig('Linear Regression')
```

## Building the Model

```python
dataset.columns
```

```python
y = dataset['Yearly Amount Spent']
```

```python
X = dataset[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

## Making Predictions

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test, predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

## Visualizing the Metrics

```python
from sklearn import metrics
```

```python
print('============ Results ============')

print('MAE: ', metrics.mean_absolute_error(y_test, predictions))
print('MSE: ', metrics.mean_squared_error(y_test, predictions))
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
sns.distplot(y_test-predictions, bins=50)
plt.savefig('Result')
```

```python
coeffecients = pd.DataFrame(lm.coef_,X.columns)
coeffecients.columns = ['Coeffecient']
coeffecients
```
